```python
#Importing necessary libraries
import json
import pandas as pd
import numpy as np
import plotly.express as px
```

```python
with open('sample_airbnb.json', 'r') as f:
    data = json.load(f)
```

```python
data[0]
```

```python
len(data)
```

```python
required_details = [
    '_id', 'listing_url', 'name', 'property_type', 'room_type',
    'accommodates', 'bedrooms', 'beds','bathrooms', 'minimum_nights', 'maximum_nights','price','cancellation_policy',
    'number_of_reviews' 
]

# Initialize an empty list to store hotel data
hotel_data = []

for hotel in data:
  try:
    # Extract required details
    hotel_info = {key: hotel.get(key) for key in required_details}

    # Add hotel info to the data list
    hotel_data.append(hotel_info)
  except json.JSONDecodeError:
    # Handle potential parsing errors (optional)
    print(f"Error parsing JSON for hotel: {hotel}")

# Create DataFrame from the list of dictionaries
hoteldf = pd.DataFrame(hotel_data)

```

```python
hoteldf 
```

```python
review_score = []

for reviews in data:
  try:
    # Extract required details
    review_info = reviews['review_scores']
    # Add hotel info to the data list
    review_score.append(review_info)
  except json.JSONDecodeError:
    # Handle potential parsing errors (optional)
    print(f"Error parsing JSON for hotel: {hotel}")

# Create DataFrame from the list of dictionaries
reviewdf = pd.DataFrame(review_score)
```

```python
reviewdf
```

```python
availability = []

for x in data:
  try:
    # Extract required details
    availability_info = {'availability_365':x['availability']['availability_365']}
    # Add hotel info to the data list
    availability.append(availability_info)
  except json.JSONDecodeError:
    # Handle potential parsing errors (optional)
    print(f"Error parsing JSON for hotel: {hotel}")

# Create DataFrame from the list of dictionaries
availabilitydf = pd.DataFrame(availability)
```

```python
availabilitydf
```

```python
#To get latitude and longitude of airbnb listings
address = []

for x in data:
  try:
    # Extract required details
    address_info = {
            "latitude": x['address']['location']['coordinates'][0],
            "longitude": x['address']['location']['coordinates'][1]
        }

    # Add hotel info to the data list
    address.append(address_info)
  except json.JSONDecodeError:
    # Handle potential parsing errors (optional)
    print(f"Error parsing JSON for hotel: {hotel}")

# Create DataFrame from the list of dictionaries
addressdf = pd.DataFrame(address)
```

```python
addressdf
```

```python
df= pd.concat([hoteldf,addressdf, reviewdf,availabilitydf], axis=1) 
```

```python
df
```

```python
df.drop(['review_scores_accuracy','review_scores_cleanliness','review_scores_checkin','review_scores_communication','review_scores_location','review_scores_value'],axis=1,inplace = True)
```

```python
df.info()
```

```python
#To check null values
df.isnull().sum()
```

```python
#Checking the duplicates
duplicates = df.duplicated(subset = '_id', keep = False)
print(duplicates.sum())
```

```python
# Impute missing data

is_rated = np.where(df['review_scores_rating'].isna() == True, "Not rated", "Rated")
df['is_rated'] = is_rated
df= df.fillna({'review_scores_rating':0,
                       })
```

```python
#Creating a column to know if rooms_breakdown is provided 
df['rooms_breakdown'] = df[['beds', 'bedrooms', 'bathrooms']].notnull().all(axis=1).astype(int)

# Fill null values with 0
df.fillna(0, inplace=True)
```
